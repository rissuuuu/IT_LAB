```python
import numpy as np
m=0
c=0
n=len(x)
# x=x.flatten()
L=0.01
reps=1000
listoferror=[]
for i in range(reps):
    y_pred=m*x+c
    d_m=(-2/n)*sum(x*(y-y_pred))
    d_c=(-2/n)*sum(y-y_pred)
    m=m-L*d_m
    c=c-L*d_c
    
    errors=np.square(np.subtract(y,y_pred))
    err=np.mean(errors)
    listoferror.append(err)
print(m,c)
print(listoferror)
```

```python
pl.figure(figsize=(10,7))
y_pred=m*x+c
pl.scatter(x,y,c=y)
pl.plot(x,y_pred,c="r")
pl.show()
len(listoferror)
```
